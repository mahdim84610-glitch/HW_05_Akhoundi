```python
import pandas as pd
import numpy as np
df = pd.read_csv("iris_dataset.csv")
features = ["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"]
X_train = df.iloc[:-1][features].values
y_train = df.iloc[:-1]['target'].values
X_test = df.iloc[-1][features].values
y_test = df.iloc[-1]['target']
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))
distances = [euclidean_distance(X_test, x) for x in X_train]
k = 3
nearest_indices = np.argsort(distances)[:k]
nearest_labels = y_train[nearest_indices]
from collections import Counter
prediction_raw = Counter(nearest_labels).most_common(1)[0][0]
print("Prediction without normalization:", prediction_raw)
print("Actual label:", y_test)
mu = X_train.mean(axis=0)
sigma = X_train.std(axis=0)
X_train_norm = (X_train - mu) / sigma
X_test_norm = (X_test - mu) / sigma
distances_norm = [euclidean_distance(X_test_norm, x) for x in X_train_norm]
nearest_indices_norm = np.argsort(distances_norm)[:k]
nearest_labels_norm = y_train[nearest_indices_norm]
prediction_norm = Counter(nearest_labels_norm).most_common(1)[0][0]
print("Prediction with normalization:", prediction_norm)
print("Actual label:", y_test)
accuracy_raw = int(prediction_raw == y_test)
accuracy_norm = int(prediction_norm == y_test)
print("Accuracy without normalization:", accuracy_raw)
print("Accuracy with normalization:", accuracy_norm)
```
