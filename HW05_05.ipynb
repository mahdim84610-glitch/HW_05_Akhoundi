```python
import pandas as pd
import numpy as np
df = pd.read_csv("iris_dataset.csv")
print(df.head())
X = df[['sepal length (cm)', 'sepal width (cm)']]
y = df['target']
Ptest = X.iloc[-1]
y_test = y.iloc[-1]
X_train = X.iloc[:-1]
y_train = y.iloc[:-1]
distances = np.sqrt(((X_train - Ptest)**2).sum(axis=1))
nearest_index = distances.idxmin()
predicted_class_k1 = y_train[nearest_index]
print("Real class:", y_test)
print("Predicted class k=1:", predicted_class_k1)
k = 3
nearest_indices = distances.nsmallest(k).index
nearest_classes = y_train[nearest_indices]
predicted_class_k3 = nearest_classes.mode()[0]
print("Predicted class k=3:", predicted_class_k3)
```
